```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 2))

# Set x-axis to log scale
ax.set_xscale('log')

# Set x-axis limits
ax.set_xlim(10**-2, 10**12)

# Set x-axis ticks at powers of 10
ticks = [10**i for i in range(-2, 13)]  # 10^-1 to 10^12
ax.set_xticks(ticks)

# Format tick labels to show as powers of 10
ax.set_xticklabels([f'$10^{{{i}}}$' for i in range(-2, 13)])

# Set x-axis label
ax.set_xlabel(r'Log($M_{\odot}$)', fontsize=12, fontweight='bold')

# Remove y-axis since we only need the x-axis scale
ax.set_yticks([])
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Keep only the bottom spine (x-axis)
ax.spines['bottom'].set_linewidth(1.5)


# Adjust layout to minimize empty space
plt.tight_layout()

# Optional: Save the figure
plt.savefig('solar_mass_scale.png', bbox_inches='tight')

# Display the plot
plt.show()
```
