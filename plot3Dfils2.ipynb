```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import mpl_toolkits.mplot3d.art3d as art3d
import filament_functions2 as fils 
import read_fils as rf
```

```python
nsig=10
name='cosmoDC2_v1.1.4_small_with_photozs_v1.TPZ.hpix9816.stamp.67.25.67.75.-37.5.-37.zslice.photo.0.6.0.8.curved.historymse3D.disperse'
path_to_skel = f'../Catalogs/{name}_nsig{nsig}_final.S002.BRK.a.NDskl'
realname='cosmoDC2_v1.1.4_small_with_photozs_v1.TPZ.hpix9816.stamp.67.25.67.75.-37.5.-37.zslice.real.0.6.0.8.3D.disperse'
```

```python
filaments=rf.ReadFilament(path_to_skel)
```

```python
fildict = filaments.filament_dict['filaments']
filnum=len(fildict)
```

```python
fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection='3d')
ax.set_position([0.05, 0.05, 0.75, 0.9])
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([2000,2300])
for i in range(filnum):
    fils.plot(i,fildict,ax=ax)
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Coordinate')
plt.draw()
plt.savefig(f'./FilPics/3D{name}.png')
```

```python
galaxies=pd.read_csv(f'../Catalogs/{name}.txt', sep='\t') 
```

```python
galaxiesplot=galaxies.query("px<10 and px>-10 and py<10 and py>-10 and pz>2000 and pz<2200")
```

```python
fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect((1, 1, 1.5))
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([2000,2200])
for i in range(filnum):
    fils.plot(i,fildict,ax=ax,colorfil="black")
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Coordinate')
ax.set_title('MSE galaxies with MSE history fils')
ax.scatter(galaxiesplot.px,galaxiesplot.py,galaxiesplot.pz,marker='.',alpha=0.1)
plt.draw()
plt.savefig(f'./FilPics/3D{name}_withgalaxies.png')
```
