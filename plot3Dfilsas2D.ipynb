```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import mpl_toolkits.mplot3d.art3d as art3d
```

```python
import filament_functions as fils 
import read_fils as rf
```

```python
nsig=10
name='cosmoDC2_v1.1.4_small_with_photozs_v1.TPZ.hpix9816.stamp.67.25.67.75.-37.5.-37.zslice.photo.0.6.0.8.curved.mse3D.disperse'
```

```python
path_to_skel = f'../Catalogs/{name}_nsig{nsig}_final.S002.BRK.a.NDskl'
```

```python
filaments=rf.ReadFilament(path_to_skel)
#down to 7310 filaments at nsig 10 but that's still way too many
```

```python
fildict = filaments.filament_dict['filaments']
```

```python
filnum=len(fildict)
```

```python
ax = plt.gca()
for i in range(filnum):
    fils.plot_3Das2D(i,fildict,axes='xy')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
ax.set_aspect('equal', adjustable='box')
tick_values = np.linspace(-10,10,9)
plt.xticks(tick_values)
plt.yticks(tick_values)
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
plt.draw()
plt.savefig(f'./FilPics/{name}_nsig{nsig}.png')
```
